```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# read the image file
image = cv2.imread('12_spur_09.jpg')

# binarization based on simple thresholding
bw = cv2.threshold(image, 70, 255, cv2.THRESH_BINARY)
(B, G, R) = cv2.split(bw[1])
new_image = cv2.merge([G, G, G])

# display images
plt.figure()
plt.imshow(image)
plt.xlabel("Input image")

plt.figure()
plt.imshow(new_image)
plt.xlabel("Output image")
```

```python
from PIL import Image

img = cv2.cvtColor(new_image, cv2.COLOR_BGR2RGB)
im = Image.fromarray(img)

width, height = im.size

# number of rows and columns
print(width//640+1, height//640+1)

images = []
# i - rows 
# j - columns
for i in range(height//640+1):
  for j in range(width//640+1):
    images.append(im.crop((j*640, i*640, (j*640)+640, (i*640)+640)))

for i in range(len(images)):
  plt.subplot(height//640+1, width//640+1, i+1)
  plt.imshow(images[i])
```
